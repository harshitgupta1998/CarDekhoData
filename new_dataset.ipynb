```python
import pandas as pd
import numpy as np
df= pd.read_csv(r"C:/Users/harshit/Desktop/CarDekho/DataSet.csv")
```

```python
final_df=df
df=pd.get_dummies(df,drop_first=True)
```

```python
df.shape
df['selling_price']=df['selling_price'].map(lambda x:x/100000)
```

```python
df.head()
```

```python
df['current_year']=2020
df['total_used_year']=df['current_year']-df['year']
df.drop(['year'],axis=1,inplace=True)
df.drop(['current_year'],axis=1,inplace=True)


```

```python
import statistics
km_driven_mean=np.mean(df['km_driven'])
km_driven_svd=statistics.stdev(df['km_driven'])
df['km_driven']=df['km_driven'].map(lambda x:(x-km_driven_mean)/km_driven_svd)
#performing feature scaling on the data set mileage,engine, and	max_power
mileage_mean=np.mean(df['mileage'])
mileage_svd=statistics.stdev(df['mileage'])
df['mileage']=df['mileage'].map(lambda x:(x-mileage_mean)/mileage_svd)

df['engine'].fillna(np.mean(df['engine']), inplace=True)
engine_mean=np.mean(df['engine'])
engine_svd=statistics.stdev(df['engine'])
df['engine']=df['engine'].map(lambda x:(x-engine_mean)/engine_svd)


df['max_power'].fillna(np.mean(df['max_power']), inplace=True)
max_power_mean=np.mean(df['max_power'])
max_power_svd=statistics.stdev(df['max_power'])
df['max_power']=df['max_power'].map(lambda x:(x-max_power_mean)/max_power_svd)
```
